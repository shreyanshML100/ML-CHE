```python
import numpy as np
import matplotlib.pyplot as plt
import linearRegression as lr
```

```python
x_train=np.array([2.5, 4.7, 5.2, 7.3, 9.5, 11.5])
y_train=np.array([5.21, 7.70, 8.30, 11, 14.5, 15])
x_train=x_train.reshape((-1,1))
y_train=y_train.reshape((-1,1))
a=np.ones((x_train.shape[0],1))
x_train=np.append(a,x_train,axis=1)
```

```python

theta3,J3=lr.fitGD(x_train,y_train,0.01,1,3,15)
print(theta3)
plt.scatter(x_train[:,1],y_train,marker='x')
plt.plot(x_train,x_train*theta3[1]+theta3[0],'r')
plt.title("scatter vs predicted line")
plt.show()
```

```python
theta2,J2=lr.fitGD(x_train,y_train,0.01,1,2,15)
plt.scatter(x_train[:,1],y_train,marker='x')
plt.plot(x_train,x_train*theta2[1]+theta2[0],'r')
plt.title("scatter vs predicted line")
plt.show()
```

```python
theta1,J1=lr.fitGD(x_train,y_train,0.01,1,1,15)
plt.scatter(x_train[:,1],y_train,marker='x')
plt.plot(x_train,x_train*theta1[1]+theta1[0],'r')
plt.title("scatter vs predicted line")
plt.show()
```

```python
theta_nor=lr.fitNormal(x_train,y_train)
# print(lr.costFun(theta_nor,x_train,y_train))
```

```python
x_test=np.array([3.5, 5, 6, 8, 10 ])
x_test=x_test.reshape((-1,1))
a=np.ones((x_test.shape[0],1))
x_test=np.append(a,x_test,axis=1)
y_test1=np.matmul(x_test,theta1)
print("linear regression1")
print(y_test1)
y_test_nor=np.matmul(x_test,theta_nor)
print("normal matrix")
print(y_test_nor)
```

```python
y_test_ac=np.array([6.25, 8.2, 9.5, 12.1, 14.7])
print("lr1 cost=",lr.costFun(theta1,x_test,y_test_ac))
print("nor cost=",lr.costFun(theta_nor,x_test,y_test_ac))
plt.plot(J1,c='r')

plt.axhline(y=lr.costFun(theta1,x_test,y_test_ac),color='b')
plt.show()
```
