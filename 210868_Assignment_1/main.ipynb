```python
import numpy as np
import linearRegression as lr
import matplotlib.pyplot as plt
```

```python
x=np.array([[2.5],[4.7],[5.2],[7.3],[9.5],[11.5]])
y=np.array([[5.21],[7.7],[8.3],[11],[14.5],[15]])
```

```python
plt.scatter(x,y,marker='x',c='r')
plt.show()
```

```python
theta=lr.fitGD(x,y,0.01,0,0,1000)
```

```python
b=theta[0]
w=theta[1:]
print("Parameters obtained corresponding to minimum cost{}".format(theta))
y_predict=np.dot(x,w)+b
plt.scatter(x,y,marker='x',c='r')
plt.scatter(x,y_predict,marker='o',c='b')
x_plot=np.arange(np.min(x),np.max(x)+1)
y_plot=w*x_plot+b
plt.plot(x_plot,y_plot)
```

```python
X_test=np.array([[3.5],[5],[6],[8],[10]])
y_test=np.dot(X_test,w)+b
plt.scatter(X_test,y_test,marker='x',c='g')
print(np.c_[X_test,y_test])
plt.scatter(x,y,marker='x',c='r')
plt.scatter(x,y_predict,marker='o',c='b')
x_plot=np.arange(np.min(x),np.max(x)+1)
y_plot=w*x_plot+b
plt.plot(x_plot,y_plot)
```
